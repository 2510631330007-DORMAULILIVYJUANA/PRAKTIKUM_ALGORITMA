```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8
v0 = 0
h0 = 100
t = np.arange(0, 11, 1) # 0 - 10 detik, step 1

s = h0 - (v0*t + 0.5*g*t**2)
v = v0 + g*t
a = g*np.ones_like(t)

plt.figure()
plt.plot(t, s, 'o-', label="Posisi (m)")
plt.plot(t, v, 's-', label="Kecepatan (m/s)")
plt.plot(t, a, '^-', label="Percepatan (m/s**2)")
plt.title("Kasus 1 - Standar jatuh bebas")
plt.xlabel("waktu (detik)")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 1")
print("Posisi maksimum:", max(s))
print("Kecepatan maksimum,", max(v))
```
